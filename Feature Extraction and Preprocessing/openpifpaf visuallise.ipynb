```python
import torch 
import numpy as np 
import openpifpaf
import cv2
from PIL import Image
from pandas import DataFrame
from sklearn.preprocessing import MinMaxScaler


# In[61]:

openpifpaf.show.Canvas.show = True
openpifpaf.show.Canvas.image_min_dpi = 200
predictor = openpifpaf.Predictor(checkpoint='shufflenetv2k30')
scaler=MinMaxScaler()
```

```python
img=cv2.imread('Analysis/occlusion/13.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_pil = Image.fromarray(img)

predictions, gt_anns, image_meta = predictor.pil_image(img_pil)
```

```python


annotation_painter = openpifpaf.show.AnnotationPainter()
with openpifpaf.show.image_canvas(img_pil) as ax:
    annotation_painter.annotations(ax, predictions,subtexts=' ',texts=' ')


```
